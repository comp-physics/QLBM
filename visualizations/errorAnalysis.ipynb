```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
i = '80'
classicalStream = np.loadtxt("classRes/stream"+i+".csv", delimiter=",")
classicalVort = np.loadtxt("classRes/vort"+i+".csv", delimiter=",")

twoCircStream = np.loadtxt("twoCirc/stream"+i+".csv", delimiter=",")
twoCircVort = np.loadtxt("twoCirc/vort"+i+".csv", delimiter=",")

oneCircStream = np.loadtxt("oneCirc/stream"+i+".csv", delimiter=",")
oneCircVort = np.loadtxt("oneCirc/vort"+i+".csv", delimiter=",")
```

```python
sns.heatmap(abs((twoCircVort - classicalVort)/np.mean(classicalVort)))
```

```python
sns.heatmap(abs((oneCircVort - classicalVort)/np.mean(classicalVort)))
```

```python
sns.heatmap(abs((twoCircStream - classicalStream)/np.mean(classicalStream)))
```

```python
sns.heatmap(abs((oneCircStream - classicalStream)/np.mean(classicalStream)))
```
