```python
import matplotlib.pyplot as plt 
import math
import numpy as np
```

```python
low = 1
high = 15
x = list(range(low, high + 1))
```

```python
one_circuit_num_qubits = []

for i in range(low, high + 1):
    M = int(math.pow(2, i)) # lattices

    dim = 2
    dirs = 5

    nlat = int(np.ceil(np.log2(M)))
    nlinks = int(np.ceil(np.log2(dirs)))

    n = nlat+nlat+nlinks+1+1 # number qubits

    one_circuit_num_qubits.append(n)
```

```python
two_circuit_num_qubits = []

for i in range(low, high + 1):
    M = int(math.pow(2, i)) # lattices

    dim = 2
    dirs = 5

    nlat = int(np.ceil(np.log2(M)))
    nlinks = int(np.ceil(np.log2(dirs)))

    n = nlinks+dim*nlat+1 # number qubits

    two_circuit_num_qubits.append(n)
```

```python
plt.scatter(x, one_circuit_num_qubits, label="One circuit qubit count")
plt.scatter(x, two_circuit_num_qubits, label="Two circuit qubit count")
plt.axhline(y = 30, color = 'r', linestyle = '-') 
plt.axvline(x = 12, color = 'r', linestyle = '-') 
plt.xlabel("2^x number lattice points")
plt.ylabel("Qubit Count")

plt.legend()
plt.show()
```
